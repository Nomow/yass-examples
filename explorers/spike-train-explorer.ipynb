```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 10)

import os.path as path

from yass.explore import SpikeTrainExplorer, RecordingExplorer
```

```python
ROOT = path.join(path.expanduser('~'), 'data/yass')
path_to_data = path.join(ROOT, 'ej49_data1_set1.bin')
path_to_geom = path.join(ROOT, 'ej49_geometry1.txt')
path_to_spike_train = path.join(ROOT, 'ej49_spikeTrain1_1.csv')
```

```python
rce = RecordingExplorer(path_to_data,
                        path_to_geom,
                        spike_size=15,
                        neighbor_radius=70,
                        dtype='int16',
                        n_channels=49,
                        data_format='long')

spe = SpikeTrainExplorer(path_to_spike_train, rce)
```

```python
spe.plot_templates(group_ids=[0, 1, 2])
```

```python
rce.plot_waveform(time=50, channels=[0, 1])
```
